- title에 중복이 12개 있다

```python
title = dataset['validation']['title']
set_title = list(set(dataset['validation']['title']))

duplicate_title = {}

for st in set_title:
    cnt = 0
    tmp = 0
    for k, t in enumerate(title):
        if st == t:
            cnt += 1
            duplicate_title[k] = t
            tmp = k
    if cnt < 2:
        duplicate_title.pop(tmp)
                
                
        
```

```python
dataset_ids_ans = {}
question_ids = {}
context_ids = {}

for i in range(80,120):
    dataset_ids_ans[dataset['validation']['id'][i]] = ''.join(dataset['validation']['answers'][i]['text'])
    question_ids[dataset['validation']['id'][i]] = ''.join(dataset['validation']['question'][i])
    context_ids[dataset['validation']['id'][i]] = dataset['validation']['context'][i]

print(len(dataset_ids_ans))
print(len(question_ids))
print(len(context_ids))
dataset_ids_ans
```

```python
question_ids['mrc-0-001486']
```

```python
cnt = 0
for pid, pans in predict_json.items():
    for ids, tex in dataset_ids_ans.items():
        if pid == ids:
            if pans == tex or tex in pans or pans in tex:
                continue
            else:
                cnt+=1
                print('\n'+'='*90+'\n')
                print(f'질문 : {question_ids[ids]}')
                print(f'predict : {pans:30s} \t  GT : {tex:20s}\n')
                print(f'{context_ids[ids]}')
print(cnt)
```
